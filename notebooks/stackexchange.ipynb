## Code

The code below uses the inbuilt corpus specific for Stack Exchange data to process the archive file and extract the posts, comments and user profiles from the downloaded archive and prepare it for further analysis.

````python
# Show the files that have been downloaded already
import glob

for path in glob.glob('data/*7z'):
    print(path)
````

````python
from hyperreal import corpus

# Create the corpus object: this is what Hyperreal uses to retrieve,
# transform and display documents in your collection
sx_corpus = corpus.StackExchangeCorpus('working/sx_data.db')

# You can pass multiple files in here and they will all be processed 
# in place: if the site was already inserted, it will be replaced with
# the new data. 
sx_corpus.replace_sites_data('data/academia.meta.stackexchange.com.7z')
````

# Working with a Corpus

A corpus does a few fundamental things:

- Identify documents with a unique key using the `keys` method
- Retrieve arbitrary documents matching given keys using the `docs` method
- Transform documents using the `doc_to_*` methods
- Extract `features` from documents for indexing and text analytics using the `doc_to_features` method

Most of the time you don't need to use these directly, but knowing about these will help if you want to customise your own corpora.

````python
from IPython.display import display, HTML

# the .keys() method retrieves all the keys for documents in the collection.
doc_keys = list(sx_corpus.keys())
# The keys for the first 10 documents in the collection
print("Keys for the first ten documents:", doc_keys[:10])

# Use the corpus to show different documents in the collection
# Note that the `docs` method retrieves documents one at a time, so you can
# deal with very large collections of documents without needing lots of
# RAM.
for key, doc in sx_corpus.docs(doc_keys[:3]):
    print(f"Document matching key {key}")
    display(HTML(sx_corpus.doc_to_html(doc)))
````

# Phase 2: Indexing and Querying the Corpus

Having created a corpus for our selected Stack Exchange sites, we can proceed to indexing that site's data. Creating an index is what allows querying and retrieve documents by the content included in that document, such as documents including the word `cat`. An index in Hyperreal is an [Inverted Index](https://en.wikipedia.org/wiki/Inverted_index), commonly used for full text search of documents.


````python
from hyperreal import index

# Create the index on the sx_corpus we created earlier
sx_idx = index.Index('working/sx_data_index.db', corpus=sx_corpus)

# Index the sx corpus data
sx_idx.rebuild()
````
